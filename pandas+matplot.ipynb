```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
df=pd.read_csv("gas_prices.csv")
df
```

```python

plt.figure(figsize=(8,6))

# plt.plot(df.Year,df.USA, label='USA',marker='o')
# plt.plot(df.Year,df['South Korea'], label='South Korea',marker='o')
# plt.plot(df.Year,df.Canada, label='Canada',marker='o')
for country in df.columns[1:]:
    plt.plot(df.Year,df[country], label=country ,marker='.')

plt.title('Gas-Prices over Time in USD')
plt.legend(title='COUNTRY', bbox_to_anchor=(1,1))
plt.xlabel('Year')
plt.ylabel('Price(in USD)')
plt.xticks(df.Year[::3])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.plot(df.Year,df.USA, label='USA',marker='.')
plt.plot(df.Year,df['South Korea'], label='South Korea',marker='.')
plt.plot(df.Year,df.Canada, label='Canada',marker='.')

plt.title('Gas-Prices over Time in USD', fontdict={'fontweight':'bold', 'fontsize':18, 'fontstyle':'italic'})
plt.legend(title='COUNTRY', bbox_to_anchor=(1,1))
plt.xlabel('Year',fontdict={'fontweight':'bold', 'fontsize':14})
plt.ylabel('Price(in USD)',fontdict={'fontweight':'bold', 'fontsize':14})
plt.xticks(df.Year[::3].tolist()+[2011])
plt.grid()
plt.savefig('Gas-Price.png', dpi=300)
plt.show()
```
