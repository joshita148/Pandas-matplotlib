```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
x1 =[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
x2=np.array([i for i in range(0,40,2)])
y2=2*x2

plt.plot(x1,y1,color='red', marker='+', linewidth=0.5, label='(x^2)+1')
plt.plot(x2,y2,color='green',marker='^',linewidth=2.5, label ='2 X x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('demo')
plt.grid()
plt.text(0,1,'intersection')

plt.legend()
plt.show()
```

```python
year=[]
with open("amazon_jobs_dataset.csv",encoding="utf8") as f:
    file_data=csv.DictReader(f)
    for row in file_data:
        if ('java' in row['BASIC QUALIFICATIONS']) or ('Java' in row['BASIC QUALIFICATIONS']):
            year.append(row['Posting_date'].split(', ')[1])
df=pd.DataFrame(year)
df.columns=['year']
a=df.year.value_counts()
x=a.index
y=a

plt.scatter(x,y)
plt.show()

for i in sorted(x):
    print(i," ",a.loc[i])
```

```python
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
colors=['cyan','pink','violet','hotpink','m','orange']
explode=(0,0.1,0,0,0,0)


plt.pie(sold, colors=colors, labels=company, explode=explode, autopct='%.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.legend(title="company")
plt.show()

x=np.array(sold)
a=x/x.sum()*100

for i in range(len(a)):
    print(company[i]," ", format(a[i],'.1f'))
```

```python
month=[]
df=pd.read_csv("amazon_jobs_dataset.csv")

for i in df.Posting_date:
    month.append(i.split(" ")[0])


df['month']=month
a=df.month.value_counts()
x=a.index
y=a

plt.bar(x,y)
plt.xticks(rotation=45)
plt.show()

for i in range(len(x)):
    print(x[i], " ", y[i])
```
